```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_fwf('/content/next_word_prediction.txt')
df.head(10)
```

```python
df['Chapter 1'][199]
```

```python
lines= []
for i in df['Chapter 1']:
  lines.append(i)

lines[:10]
```

```python
len(lines)
```

```python
# Convert List o string
data= ""
for i in lines:
  data=' '.join(lines)

len(data)
```

```python
data[:2000]
```

```python
import re
def preprocess(review):
    review= review.lower()
    review= re.sub('[^a-z]',' ',review)
    words= review.split()
    return ' '.join(words)
```

```python
data_processed= preprocess(data)
```

```python
data_processed[:2000]
```

```python
len(data_processed)
```

```python
#df_processed=pd.DataFrame(data_processed)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer= Tokenizer()
tokenizer.fit_on_texts([data_processed])
```
