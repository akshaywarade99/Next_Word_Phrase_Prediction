```python
import pandas as pd
import numpy as np
```

```python

df = pd.read_fwf('/content/next_word_prediction.txt')
df.head(-10)
```

```python
df['Chapter 1']
```

```python
lines= []
for i in df['Chapter 1']:
  lines.append(i)

lines[:1]
```

```python
len(lines)
```

```python
# data_processed_seq = tokenizer.texts_to_sequences(data_processed)
```

```python
# Convert List o string
data= ""
for i in lines:
  data=' '.join(lines)

len(data)
```

```python
data[:-20]
```

```python
import re
def preprocess(review):
    review= review.lower()
    review= re.sub('[^a-z]',' ',review)
    words= review.split()
    return ' '.join(words)
```

```python
data_processed= preprocess(data)
```

```python
data_processed[:200]
```

```python
len(data_processed)
```

```python
#df_processed=pd.DataFrame(data_processed)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer= Tokenizer()
tokenizer.fit_on_texts([data_processed])
```

```python
len(tokenizer.index_word)
```

```python
# from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data_processed_seq = tokenizer.texts_to_sequences([data_processed])
```

```python
data_processed_seq
```

```python
len(data_processed_seq[0])
```

```python
# data_processed_seq_pad = pad_sequences(data_processed_seq)
```

```python
from tensorflow.keras.layers import LSTM, Embedding, Dense
from tensorflow.keras.models import Sequential
```

```python
sequences=[]
for i in range(3, len(data_processed_seq[0])):
  words=data_processed_seq[0][i-3:i+1]
  sequences.append(words)

len(sequences)
```

```python
import numpy as np
```

```python
data_array=np.array(sequences)
```

```python
data_array[:10]
```

```python
X=data_array[:,0:3]
y=data_array[:,3]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
text_size=len(tokenizer.word_index)+1
```

```python
y=to_categorical(y,num_classes=text_size)
y[:5]
```

```python
# model building
model=Sequential()
model.add(Embedding(text_size,10,input_length=3))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100,return_sequences=True))
model.add(LSTM(100,return_sequences=False))
model.add(Dense(100,activation='relu'))
model.add(Dense(text_size,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100,batch_size=32)
```
